# Project Setup & Data Verification

This notebook verifies the environment and performs initial exploratory data analysis (EDA) on the Chest X-Ray (Pneumonia) dataset.

```python
import os
import torch
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

# Check Device
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")
```

## 1. Dataset Overview

We load the dataset and check the class distributions across the train, val, and test splits.

```python
data_root = '../data/chest_xray'
splits = ['train', 'val', 'test']
labels = ['NORMAL', 'PNEUMONIA']

stats = []
for split in splits:
    for label in labels:
        path = os.path.join(data_root, split, label)
        if os.path.exists(path):
            count = len(os.listdir(path))
            stats.append({'split': split, 'label': label, 'count': count})

df_stats = pd.DataFrame(stats)
print(df_stats)

df_stats.pivot(index='split', columns='label', values='count').plot(kind='bar', stacked=True)
plt.title("Class Distribution Per Split")
plt.show()
```

## 2. Sample Visualization

Visualizing a random selection of 'Normal' and 'Pneumonia' cases.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for i, label in enumerate(labels):
    path = os.path.join(data_root, 'train', label)
    img_name = os.listdir(path)[0]
    img = Image.open(os.path.join(path, img_name)).convert('RGB')
    axes[i].imshow(img)
    axes[i].set_title(f"Sample: {label}")
    axes[i].axis('off')

plt.show()
```
